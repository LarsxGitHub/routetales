```python
def load_data():
  data = dict()
  for fn in tqdm.tqdm(glob.glob("../data/fwd_diff/*/pfxfwdres_*.pkl")): 
      with open(fn, 'rb') as fh:
        d = pickle.load(fh)
        data[d.end_date] = d
  return data
```

```python
def retrieve_characteristic_timeline(data, aslevel: bool, window: str, cname: str):
  """ cname options:

  cnt_pfx
  has_no_direct_msp_coverage
  has_partial_direct_msp_coverage
  has_full_direct_msp_coverage
  has_no_msp_coverage
  has_partial_msp_coverage
  has_full_msp_coverage
  has_sibling
  has_direct_less_specific
  has_less_specific

  Window options:
  Week-IPv6
  Month-IPv6
  Year-IPv6
  Week-IPv4
  Month-IPv4
  Year-IPv4
  """
  xs, gains, mains, losses = [], [], [], []
  for date in sorted(data.keys()):
    xs.append(date)
    if aslevel:
      cdiff = getattr(data[date].results[window].asn_diff, cname)
    else:
      cdiff = getattr(data[date].results[window].pfx_diff.class_totals, cname)

    gains.append(cdiff.gain)
    mains.append(cdiff.main)
    losses.append(cdiff.loss)
  return xs, gains, mains, losses
```

```python
d = load_data()
```

```python
xs, gains, mains, losses = retrieve_characteristic_timeline(d, False, "Year-IPv4", "cnt_pfx")
print(xs)
print(gains)
print(mains)
print(losses)
```
